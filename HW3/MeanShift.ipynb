```python
import numpy as np
```

```python
def mean_shift(X, y, K, R, threshold):
    '''input:
        X: input data, the shape of X is (N, D)
        y: the label of X, use to calculate the accuracy
        K_centers: the K centers, shape is (K, D)
        R: R=80, for normalized X range is [0,1], radius
       output:
        K centers, shape of K is (K, D)
    '''
    import random
    N, D = X.shape
    
    count = -1
    K_centers = np.zeros((K, D)) + 1000
    y_pred = np.zeros((N,1), np.int)
    visited = np.zeros((N,1), np.int)
    unvisit_idx = np.argwhere(visited==0)
#     print unvisit_idx.shape, unvisit_idx[0]
    while unvisit_idx.shape[0] > 0:
        # choose a center from X which not visited yet, randomly
#         unvisit_idx = np.argwhere(visited==0)
        center_idx = unvisit_idx[random.randint(0, unvisit_idx.shape[0] - 1)]
        print 'new center'
        print center_idx[0]
        center =  X[center_idx[0]].reshape(1,-1)
        while 1:
#             print 'while loop'
            # calculate the distance of center to all Xi
            dis = np.linalg.norm(X-center, axis=1)
#             print dis.shape
#             print dis
            # get the indices that belongs to the interesting area
            idx = np.argwhere(dis<R)
            # mark visited
#             print idx.shape
#             print idx
            visited[idx] = 1
            
#             # label these point as the value of count
#             y_perd[idx] = count
            # calculate the Mean Shift Vector
            Mh = np.mean((X[idx] - center), axis=0)
            # if shift distance is small enough, then this shift finish
            if np.linalg.norm(Mh) < threshold:
                break
            # update the center
            center += Mh
        
        # all visited indices each shift
        idx = np.argwhere(visited==1)
        print 'each cluster, number of point'
        print idx.shape
        # Union the center
        Union_center = -1
        for i in range(count):
#             print 'K_centers: '+str(K_centers[i].shape)
            center_dis = np.linalg.norm(K_centers[i] - center)
#             print 'center_dis :'+ str(center_dis)
            # if there are close, then merge them
            if center_dis < R/2:
#                 print 'merge : '+str(i)
#                 y_pred[idx] = i
                Union_center = i
                break
        # no need to merge
        if Union_center == -1:
            # add a new center
            count += 1
            K_centers[count] = center
            y_pred[idx[:,0].reshape(-1)] = count
        else:
            # update the center after merge
            K_centers[Union_center] = 0.5*(K_centers[Union_center] + center)
        visited[idx[:,0].reshape(-1)] += 1
        unvisit_idx = np.argwhere(visited==0)
        print 'left point not visited'
        print unvisit_idx.shape
        
    y_pred_label = np.zeros((N,1), np.int)
    for i in range(count):
#         print i
        idx = np.argwhere(y_pred == i)
        if idx.shape[0] == 0:
            break
#         print idx
        count_label = np.bincount(y[idx[:,0].reshape(-1)].reshape(-1))
        y_pred_label[idx[:,0].reshape(-1)] = np.argmax(count_label)
#     print count
#     print y_pred_label.reshape(-1)
#     print y.reshape(-1)
    acc = np.mean(y_pred_label.reshape(-1)==y.reshape(-1))
    print 'accuracy is : ' +str(acc)
    return y_pred_label
```
